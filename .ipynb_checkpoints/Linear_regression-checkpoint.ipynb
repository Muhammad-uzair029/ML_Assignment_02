```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel(r"C:\Users\HP\Assignment_2\trainning_Dataset.xlsx")
df = df.dropna(axis='columns', how='all')
df_test = pd.read_excel(r"C:\Users\HP\Assignment_2\test_Dataset.xlsx")


df_test = df_test.dropna(axis='columns', how='all')
# Drop the null rows
df_test= df_test.dropna()
# df_test.head()
df_test[:150]
```

```python
#creating a dtaframe 
df_x = pd.DataFrame(df, columns = ['cylinders', 'acceleration','model year'])
```

```python
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split

# Extract 2    Row from test Dataset for prediction
X_train, X_test, y_train, y_test = train_test_split(df_x, df.mpg, test_size=0.20,random_state=80)
reg_model_one = linear_model.LinearRegression()
reg_model_one.fit(X_train,y_train)
reg_model_one_prediction=reg_model_one.predict([[8,12,70]])

accuracy_model_one = reg_model_one.score(X_test,y_test)


#Extract  100 Row from test Dataset for prediction

X_train, X_test, y_train, y_test = train_test_split(df_x, df.mpg, test_size=0.20,random_state=160)
reg_model_two = linear_model.LinearRegression()
reg_model_two.fit(X_train,y_train)
reg_model_two_prediction=reg_model_two.predict([[4,14.7,79]])

accuracy_model_two = reg_model_two.score(X_test,y_test)


# Extract 130 Row from test Dataset for prediction
X_train, X_test, y_train, y_test = train_test_split(df_x, df.mpg, test_size=0.20,random_state=241)
reg_model_three = linear_model.LinearRegression()
reg_model_three.fit(X_train,y_train)
reg_model_three_prediction=reg_model_three.predict([[4,24.6,82]])

accuracy_model_three = reg_model_three.score(X_test,y_test)


print('Prediction of First Subset Dataset',reg_model_one_prediction)
print('Prediction of Second Subset Dataset',reg_model_two_prediction)
print('Prediction of Third Subset Dataset',reg_model_three_prediction)


print('Accuracy of First Subset Dataset',accuracy_model_one*100,'%')
print('Accuracy of Second Subset Dataset',accuracy_model_two*100,'%')
print('Accuracy of Third Subset Dataset',accuracy_model_three*100,'%')
```

```python
%matplotlib inline
plt.xlabel('Prediction of Model')
plt.ylabel('Accuracy Of Model')

Y_Axis=[reg_model_one_prediction,reg_model_two_prediction,reg_model_three_prediction]
X_Axis=[accuracy_model_one*100,accuracy_model_two*100,accuracy_model_three*100]

plt.plot(X_Axis,Y_Axis,color='red')
plt
```
